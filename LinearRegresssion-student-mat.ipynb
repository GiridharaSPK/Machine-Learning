LINEAR REGRESSION (WITH FEATURE SELECTION)
To predict the values of G3 the most corelated features are G1 and G2 with corelation values 0.8 and 0.9


```python
x = Data.iloc[:,[30, 31]]
y = Data.iloc[:,[32]]
```

```python
# splitting x and y into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=3)
print("Training subset: ", x_train.shape, y_train.shape)
print("Testing subset: ", x_test.shape, y_test.shape)
```

```python
#linear regression object : lin_reg
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
# Make predictions using the testing set
y_pred = lin_reg.predict(x_test)
print("Predictions size: ", y_pred.shape)
print("Predictions : \n", y_pred)
```

```python

print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_pred))
#print('Fscore: %.2f'% f1_score(y_test, y_pred, average=None))
```

```python
clf_y = Data['school']
clf_X = Data.drop('school',axis=1)
```

```python
cx_train, cx_test, cy_train, cy_test = train_test_split(clf_X,clf_y,test_size=0.3,random_state=3)

logi = LogisticRegression(fit_intercept=True)
logi.fit(cx_train,cy_train)

logi_regression_preds = logi.predict(cx_test)
logi_cf = confusion_matrix(cy_test,logi_regression_preds)
print("Confusion matrix of logistic regression")
print(logi_cf)
print('Fscore: ')
logi_acc= f1_score(cy_test,logi_regression_preds)
print('Fscore: ',logi_acc)
```

```python
svm_clf = svm.SVC()
svm_clf.fit(cx_train,cy_train)
svm_preds = svm_clf.predict(cx_test)
svm_cf = confusion_matrix(cy_test,svm_preds)
print("Confusion matrix of SVM")
print(svm_cf)
svm_acc=f1_score(cy_test,svm_preds)

print({'F-score':svm_acc})
```

Fscore of Logistic regression < SVM 

Hence we can say that SVM is better
